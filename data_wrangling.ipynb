# Data wrangling データ操作

# データセット読み込み ボストンの住宅価格回帰データセット

```python
from keras.datasets import boston_housing

(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

# データ表示設定

```python
import numpy as np
import pandas as pd

# ndarrayの最大表示数 配列のサイズがthresholdを超えた場合省略される
np.set_printoptions(threshold=1000)
# DataFrame表示設定
pd.options.display.max_columns = None
pd.options.display.max_rows = 100
```

# ndarrayからDataFrameへ変換

```python

df = pd.DataFrame(x_train)
```

# データ情報確認

```python
df.info()
print(df.max())
print(df.count())
```

```python
# count 平均 標準偏差 など
df.describe()
```

# 列名の変更

```python

# 以下から
# https://scikit-learn.org/stable/datasets/index.html#boston-house-prices-dataset
df = df.rename(columns={
    0: 'CRIM',
    1: 'ZN',
    2: 'INDUS',
    3: 'CHAS',
    4: 'NOX',
    5: 'RM',
    6: 'AGE',
    7: 'DIS',
    8: 'RAD',
    9: 'TAX',
    10: 'PTRATIO',
    11: 'B',
    12: 'MEDV'
})
```
